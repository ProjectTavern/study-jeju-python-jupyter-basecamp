# Dataframe

우리가 Excel에서 보듯이 2차원적인 데이터를 저장한 형태.
Series가 모이면 DataFrame이라고 생각하면 된다.

```python
data = {
        'Name': 'Dongsoo',
        'Age': 21,
        'Height': 180.2,
        'Job': 'Student'
       }
data
```

```python
columns = ['Name', 'Age', 'Height', 'Weight', 'Job']
```

```python
df = pd.DataFrame(data = data, index = [0], columns = columns)
df
```

```python
multiData = {
        'Name': ['Dongsoo', 'SeYoung'],
        'Age': [21, 31],
        'Height': [180.2, 175],
        'Weight': [58.5, 55.2],
        'Job': ['Student', 'Officer']
       }

multiData
```

```python
df = pd.DataFrame(data = multiData, index = [0, 1], columns = columns)
df
```
